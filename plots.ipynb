```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cms_ = [0, 2.5, 5, 7.5]
fvals_ = [4400, 3500, 2100, 1500]
bvals_ = [4400, 3700, 1700, 1300]
cms = np.arange(0, 8, 0.5)
fvals = np.interp(cms, cms_, fvals_)
bvals = np.interp(cms, cms_, bvals_)
```

```python
# Set Seaborn theme for better visuals
sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 6), dpi=300)
plt.plot(cms, fvals, marker='o', label="Front Values", linestyle='--', color='blue')
plt.plot(cms, bvals, marker='o', label="Back Values", linestyle='--', color='red')

# Add titles and labels
plt.title("Horizontal Sensors", fontsize=14)
plt.xlabel("cm", fontsize=14)
plt.ylabel("Values", fontsize=14)

# Customize ticks
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Add legend
plt.legend(title="Legend", fontsize=11, title_fontsize=11)

# Show grid for better readability
plt.grid(visible=True, linestyle=':', linewidth=0.5, alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
